```python
# (3) :
#HISTOGRAMS PLOT USING MATPLOTLIB


life_exp = np.array([23,45,23,55,34])
gdp_cap = np.array([234,345,456,567,678])

# Build histogram with 5 bins
plt.hist(life_exp, 5)

# Show and clean up plot
plt.show()
plt.clf()

# Build histogram with 20 bins
plt.hist(life_exp,20)

# Show and clean up again
plt.show()
plt.clf()
```

```python
# (4) :
#Customization of plots using MATPLOTLIB

# Basic scatter plot, log scale
plt.scatter(gdp_cap, life_exp)

#log scale
plt.xscale('log') 

# Strings
xlab = 'GDP per Capita [in USD]'
ylab = 'Life Expectancy [in years]'
title = 'World Development in 2007'

# Add axis labels
plt.xlabel(xlab)
plt.ylabel(ylab)


# Add title
plt.title(title)

# After customizing, display the plot
plt.show()
```

```python
# (5) :
#CUSTOMIZATION OF PLOTS USING MATPLOTLIB

# x-axis and y-axis tick labels :

# Basic scatter plot
plt.scatter(gdp_cap, life_exp)

# Previous customizations
plt.xlabel('GDP per Capita [in USD]')
plt.ylabel('Life Expectancy [in years]')
plt.title('World Development in 2007')

# Definition of tick_val and tick_lab
tick_val = [1000, 10000, 100000]

# Adapt the ticks on the x-axis
plt.xticks(tick_val)

# After customizing, display the plot
plt.show()



                              


```

```python
                                    #Double ticks on the x-axis


# Basic scatter plot
plt.scatter(gdp_cap, life_exp)

# Previous customizations
plt.xscale('log')
plt.xlabel('GDP per Capita [in USD]')
plt.ylabel('Life Expectancy [in years]')
plt.title('World Development in 2007')

# Definition of tick_val and tick_lab
tick_val = [1000, 10000, 100000]
tick_lab = ['1k', '10k', '100k']

# Adapt the ticks on the x-axis
plt.xticks(tick_val,tick_lab)

# After customizing, display the plot
plt.show()

                               
```

```python
# (6) :
#ADD (Size) to the Scatter plot function using MATPLOTLIB

pop = [23,45,23,55,34]

# Store pop as a numpy array: np_pop
np_pop = np.array(pop)

# Double np_pop
np_pop = np_pop*2

# Update: set s argument to np_pop
plt.scatter(gdp_cap, life_exp, s = np_pop)

# Previous customizations
plt.xscale('log') 
plt.xlabel('GDP per Capita [in USD]')
plt.ylabel('Life Expectancy [in years]')
plt.title('World Development in 2007')
plt.xticks([1000, 10000, 100000],['1k', '10k', '100k'])

# Display the plot
plt.show()
```

```python
#7 :
#ADD (Colors) to the Scatter plot function using MATPLOTLIB  and transparency (alpha)

#color dictionary
col = {
    'Asia':'red',
    'Europe':'green',
    'Africa':'blue',
    'Americas':'yellow',
    'Oceania':'black'
}

region_colors = [col[region] for region in col]


# Specify c for color and alpha inside plt.scatter()             &&            #alpha is for transparency

plt.scatter(x = gdp_cap, y = life_exp, s = np.array(pop) * 2, c= region_colors, alpha = 0.8)  

# Previous customizations
plt.xscale('log') 
plt.xlabel('GDP per Capita [in USD]')
plt.ylabel('Life Expectancy [in years]')
plt.title('World Development in 2007')
plt.xticks([1000,10000,100000], ['1k','10k','100k'])

# Show the plot
plt.show()
```

```python
# (8) :
#ADD (TEXT) to the Scatter plot function using MATPLOTLIB


# Scatter plot
plt.scatter(gdp_cap, life_exp, s = np.array(pop) * 2, c = region_colors, alpha = 0.9)

# Previous customizations
plt.xscale('log') 
plt.xlabel('GDP per Capita [in USD]')
plt.ylabel('Life Expectancy [in years]')
plt.title('World Development in 2007')
plt.xticks([1000,10000,100000], ['1k','10k','100k'])

# Additional customizations
plt.text(1550, 71, 'India')
plt.text(5700, 80, 'China')

# Add grid() call
plt.grid(True)

# Show the plot
plt.show()
```
