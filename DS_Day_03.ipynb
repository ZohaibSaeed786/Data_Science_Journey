<h1>Importig Libraries 

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
```

<h1>Creating a DataFrame using lists


```python

countries =  ['Pakistan','India','Afghanistan','China']
cp = ['islamabad','delhi','kabul','beijing']
pp =  ['30','130','10','140']
gd  =  ['2.14','4.22','3.25','4.42']

world = { 'Country': countries ,'Capital': cp ,'pop'   : pp ,'GDP'   : gd }


data = pd.DataFrame(world)
data
```

<h1>Filtering DataFrames</h1>

<h4>We can filter the data in a dataframe using the following syntax</h4>
<h4>By using the comparison operators and the boolean operators</h4>

<h4>Comparison Operators and Boolean operators</h4>

<h4>So we can filter it by using the loc or iloc to select the columns and rows</h4>




```python
#Suppose we want to select the countries with the gdp higher than 4.0

hg = data.loc[:,'GDP'] > str(4.0)
#This will return a boolean series with True and False values
high_gdp_countries = data[hg]

high_gdp_countries
```

<h2>One line method is : </h2>

```python
#Suppose we want to select the countries with the gdp higher than 4.0

high_gdp_countries = data[data.loc[:,'GDP'] > str(4.0)]
high_gdp_countries
```

```python
temp = data[np.logical_and( data.loc[ : , 'pop'] > '100', data.loc[:, 'pop'] < '150')]
temp
```

<h1>Loop Data Structures 1</h1>

<h4>For loop</h4>

```python
# areas list
areas = [11.25, 18.0, 20.0, 10.75, 9.50]

# Code the for loop
for x in areas:
    print(x)
```

<h4>If you want to print the index, then use enumerate() function</h4>

```python
# areas list
areas = [11.25, 18.0, 20.0, 10.75, 9.50]

# Change for loop to use enumerate() and update print()
for index,x in enumerate(areas) :
    print("room" + str(index) + ":" + str(x))
```

<h2> For Dictionaries, we use method [ .items() ] for For loop</h2>

```python
# Definition of dictionary
europe = {'spain':'madrid', 'france':'paris', 'germany':'berlin',
'norway':'oslo', 'italy':'rome', 'poland':'warsaw', 'austria':'vienna' }
          
# Iterate over europe
for k,x in europe.items():
    print("the capital of " + k + " is "+ str(x))
```

<h2>For Numpy Arrays, we use function [ np.nditer(array_name) ] function</h2>

```python
#Numpy arrays

np_height= np.array([11.25, 18.0, 20.0, 10.75, 9.50])
np_baseball= np.array([23.25, 68.0, 34.0, 64.75, 3.50])

# For loop over np_height
for x in np.nditer(np_height):
    print(str(x) + " inches")

# For loop over np_baseball
for x in np.nditer(np_baseball):
    print(str(x))
```

<h3>Loop Data Structures 2</h3>

<h4>If pandas dataframe is used, then we use [ .iterrows() ] function</h4>

```python
    
countries =  ['Pakistan','India','Afghanistan','China']
cp = ['islamabad','delhi','kabul','beijing']
pp =  ['30','130','10','140']
gd  =  ['2.14','4.22','3.25','4.42']

world = { 'Country': countries ,'Capital': cp ,'pop'   : pp ,'GDP'   : gd }


data = pd.DataFrame(world)


for lab,values in data.iterrows():
    print(lab)
    print(values)
```

<h3>if we want to add a column by using for loop, then we use [ .apply() ] function</h3>

```python
for lab,values in data.iterrows():
    data[lab, "name_length"] = len(values["Country"])
print(data)
```

<h4>if we want to add column without using for loop, then we use [ .apply() ] function</h4>
<h5>For big datasets, we use [ .apply() ] function</h5>

```python
data["name_length"] = data["Country"].apply(len)
print(data)
```

<h3>If we want to print only selected columns, then we will do this</h3>

```python
for lab, row in data.iterrows():
    print(str(lab) + ": " + str(row["Country"]))
```

```python
# Starting step
step = 50

# Roll the dice
dice = np.random.randint(1,7)

# Finish the control construct
if dice <= 2 :
    step = step - 1
elif dice>2 and dice<=5 :
    step = step +1
else :
    step = step + np.random.randint(1,7)

# Print out dice and step
print(dice, step)
```

<h2>use of max() function</h2>

```python
# Initialize random_walk
random_walk = [0]

for x in range(100):
    step = random_walk[-1]
    dice = np.random.randint(1,7)
    if dice <= 2:
        # Replace below: use max to make sure step can't go below 0
        step = max(0, step - 1)
    elif dice <= 5:
        step = step + 1
    else:
        step = step + np.random.randint(1,7)
    random_walk.append(step)
print(random_walk)
```

<h2>Use of the tanspose() function and then visualization</h2>

```python
# numpy and matplotlib imported, seed set.

# initialize and populate all_walks
all_walks = []
for i in range(5) :
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)
        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
        random_walk.append(step)
    all_walks.append(random_walk)

# Convert all_walks to NumPy array: np_aw
np_aw = np.array(all_walks)

# Plot np_aw and show
plt.plot(np_aw)
plt.show()

# Clear the figure
plt.clf()

# Transpose np_aw: np_aw_t
np_aw_t = np.transpose(np_aw)


# Plot np_aw_t and show
plt.plot(np_aw_t)
plt.show()
```

<h2>creating histogram</h2>

```python
# numpy and matplotlib imported, seed set

# Simulate random walk 500 times
all_walks = []
for i in range(500) :
    random_walk = [0]
    for x in range(100) :
        step = random_walk[-1]
        dice = np.random.randint(1,7)
        if dice <= 2:
            step = max(0, step - 1)
        elif dice <= 5:
            step = step + 1
        else:
            step = step + np.random.randint(1,7)
        if np.random.rand() <= 0.001 :
            step = 0
        random_walk.append(step)
    all_walks.append(random_walk)

# Create and plot np_aw_t
np_aw_t = np.transpose(np.array(all_walks))

# Select last row from np_aw_t: ends
ends = np_aw_t[-1,:]

# Plot histogram of ends, display plot
plt.hist(ends)
plt.show()
```
